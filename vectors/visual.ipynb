```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
plt.plot(1, 2, 'ko') # plot a black circle
plt.plot([1, 3], [2, 3], 'r--') # plot a line x-axis (0 -> 2), y-axis (0 -> 4)
plt.xlim([-4, 4]) # set the x-axis limits
plt.ylim([-4, 4]) # set the y-asix limist

plt.plot(-3, -2, 'ko')
plt.plot([-3,3],[-2, 3],'r--')

plt.plot(0,0, 'ko')
plt.plot([0,-3],[0, -2], 'r--')
plt.title('The graph title') # graph title
```

```python
_,axs = plt.subplots(2, 2, figsize=(8, 5)) # create subplots (grid) geometry (2 x 2) -> [[(0,0), (0,1)], [(1,0), (1,1)]]
axs[0,0].plot(np.random.randn(10, 5)) # line plot on the (0,0)
axs[0, 1].plot(np.random.randn(11, 4)) # line plot on the (0, 1)
axs[1, 0].imshow(np.random.randn(6, 11)) # image on the (1, 0)
axs[1, 1].imshow(np.random.randn(10, 5)) # line on the (1, 1)
```

```python
x = np.arange(-4, 5)
print(x)
y = x**2
print(y)
plt.plot(x, y, 's-')
```

```python
x = np.linspace(-4, 4, 42)
y = x**2
plt.plot(x, y, 's--')
```

```python

clockStart = time.time()
x = np.linspace(-12, 12, 42)
alpha = 1.4
beta = 2

num = alpha # numerator
den = 1 + np.exp(-beta*x) # denominator
fx = num / den
plt.plot(x, fx, 's--')
compTime = time.time() - clockStart
print(f'Computation time: {compTime} seconds')
```
