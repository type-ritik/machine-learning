```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
# Exercise 2-1
v = np.array([1,2])
w = np.array([4,-6])
vMinusW = v-w
print(vMinusW)

# now plot all three vectors
plt.figure(figsize=(6,6))

a1 = plt.arrow(0, 0, v[0], v[1], head_width=.3, width=.1, color='k', length_includes_head=True)

a2 = plt.arrow(w[0], w[1], vMinusW[0], vMinusW[1], head_width=.3, width=.1, color=[.8, .8, .8], length_includes_head=True)

a3 = plt.arrow(0, 0, w[0], w[1], head_width=.3, width=.1, color=[.5, .5,.5], length_includes_head=True)

# make the plot look a bit nicer
plt.grid(linestyle='--', linewidth=.5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([a1, a2, a3], ['v', 'w', 'v-w'])
plt.title('Vectors v, w, and v-w')
plt.savefig('./figures/Figure_02_03a.png', dpi=300)
plt.show()
```

```python
# Exercise 2-2
def norm(vec_num):
    return np.sqrt(np.sum(np.power(vec_num, 2)))

# Random GenInput0
vec = np.random.randn(3)
print(norm(vec))
print(np.linalg.norm(vec))

# Test Input1
print(norm([1, 2, 3]))
print(np.linalg.norm([1, 2, 3]))

# Test Input2
print(norm([0, 0, 1]))
print(np.linalg.norm([0, 0, 1]))
```

```python
# Exercise 2-3
def u_vec(vec_num):
    num = np.array(vec_num)
    den = np.linalg.norm(vec_num)
    if den == 0:
        return [0, 0, 0]
    return num / den



print(u_vec(np.zeros((4, 1))))
```

```python
# Exercise 2-4
def scaled_magnitude(ori_vec, des_mag):
    unit_vec = u_vec(ori_vec)
    return unit_vec * des_mag


v = [3, 4]
m = 10
mv =scaled_magnitude(v, m)
print(mv)

print(np.linalg.norm(v), np.linalg.norm(mv))
```

```python
# Exercise 2-5
vec = np.array([[9, 3, 8]])
def row_to_col_transpose(ori_vec):
    col_vec = []
    for itm in ori_vec[0]:
        col_vec.append([itm])

    return np.array(col_vec)


def col_to_row_transpose(ori_vec):
    row_vec = [[]]
    for itm in range(len(ori_vec)):
        row_vec[0].append(ori_vec[itm][0])

    return np.array(row_vec)
col_vec =row_to_col_transpose(vec)
print(f"Column Vector: {col_vec.shape}")
print()
row_vec = col_to_row_transpose(col_vec)
print(f"Row Vector: {row_vec.shape}")
```

```python
# Exercise 2-7
def dot_product(vec_a, vec_b):
    return vec_a * vec_b == vec_b * vec_a

vec_a = np.array(np.random.randn(11, 1))
vec_b = np.array(np.random.randn(11, 1))

atb = np.sum(vec_a * vec_b)
bta = np.sum(vec_b * vec_a)

print(atb - bta)
# Error for not having same dimension
print(f"{np.dot(vec_a[0], vec_b[0])} : {np.dot(vec_b[0], vec_a[0])}")
print(dot_product(vec_a, vec_b))
```

```python
# Exercise 2-8

# the vectors a, b
a = np.array([1, 2])
b = np.array([1.5, .5])

# compute beta (a*b) / (a*a) -> Transpose isn't matter
beta = np.dot(a, b) / np.dot(a, a)
print(beta)

# compute the projection vector
projvect = b - beta*a
print(projvect)

# draw the figure
plt.figure(figsize=(4,4))

# vectors
plt.arrow(0, 0, a[0], a[1], head_width=.2, width=.02, color='k', length_includes_head=True)
plt.arrow(0, 0, b[0], b[1], head_width=.2, width=.02, color='k', length_includes_head=True)

# projection vector
plt.plot([b[0], beta*a[0]], [b[1], beta*a[1]], 'k--')

# projection on a
plt.plot(beta*a[0], beta*a[1], 'ko', markerfacecolor='w', markersize=13)

# make the plot look nicer
plt.plot([-1, 2.5], [0, 0], '--', color='gray', linewidth=.5)
plt.plot([0, 0], [-1, 2.5], '--', color='gray', linewidth=.5)

# add labels
plt.text(a[0] + .1, a[1], 'a', fontweight='bold', fontsize=18)
plt.text(b[0], b[1] - .3, 'b', fontweight='bold', fontsize=18)
plt.text(beta*a[0] - .34, beta*a[1], r'', fontweight='bold', fontsize=18)
plt.text((b[0] + beta*a[0]) / 2, (b[1]+beta*a[1])/2 + .1, r'(b-a)', fontweight='bold', fontsize=18)

# some finishing touches
plt.axis('square')
plt.axis([-1, 2.5, -1, 2.5])
```

```python
# Exercise 2-9

# generate random R2 vectors
t = np.random.randn(2)
r = np.random.randn(2)

# the decomposition
t_para = r * (np.dot(t, r) / np.dot(r, r))
t_perp = t - t_para

# confirm that the two components sum to the target
print(t)
print(t_para+t_perp)

# confirm orthogonality that the product must be zero
print(np.dot(t_para, t_perp))

# draw them!
plt.figure(figsize=(6, 6))

# draw main vectors
plt.plot([0, t[0]], [0, t[1]], color='k', linewidth=3, label=r't')
plt.plot([0, r[0]], [0, r[1]], color=[.7, .7, .7], linewidth=3, label=r'r')

# draw decomposed vector components
plt.plot([0, t_para[0]], [0, t_para[1]], 'k--', linewidth=3, label=r'tp')
plt.plot([0, t_perp[0]], [0, t_perp[1]], 'k:', linewidth=3, label=r't+')

plt.axis('equal')
plt.legend()
# plt.savefig('./figures/figure_02_08.png', dpi=300)
```
