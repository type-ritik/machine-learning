```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
# Exercise 2-1
v = np.array([1,2])
w = np.array([4,-6])
vMinusW = v-w
print(vMinusW)

# now plot all three vectors
plt.figure(figsize=(6,6))

a1 = plt.arrow(0, 0, v[0], v[1], head_width=.3, width=.1, color='k', length_includes_head=True)

a2 = plt.arrow(w[0], w[1], vMinusW[0], vMinusW[1], head_width=.3, width=.1, color=[.8, .8, .8], length_includes_head=True)

a3 = plt.arrow(0, 0, w[0], w[1], head_width=.3, width=.1, color=[.5, .5,.5], length_includes_head=True)

# make the plot look a bit nicer
plt.grid(linestyle='--', linewidth=.5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([a1, a2, a3], ['v', 'w', 'v-w'])
plt.title('Vectors v, w, and v-w')
plt.savefig('Figure_02_03a.png', dpi=300)
plt.show()
```
